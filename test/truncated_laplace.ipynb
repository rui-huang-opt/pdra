# Test for truncated Laplace noise

```python
import numpy as np
import matplotlib.pyplot as plt
from pdra import TruncatedLaplace

low = -2          # Lower bound of the distribution.
high = 1          # Upper bound of the distribution.
loc = 0           # Location parameter of the distribution.
scale = 1         # Scale parameter of the distribution.
n_samples = 50000 # Number of samples to draw.
n_bins = 50       # Number of bins to plot histogram.
n_points = 1000   # Number of points to plot pdf.

tl = TruncatedLaplace(low, high, loc, scale)

samples = tl.sample(n_samples)

figure_path = "figures/truncated_laplace/"

fig, ax = plt.subplots()

ax.hist(samples, bins=n_bins, density=True)

x = np.linspace(low, high, n_points)

ax.plot(x, tl.pdf(x), "r", lw=2)

ax.set_title("Truncated Laplace Distribution")
ax.set_xlabel("x")
ax.set_ylabel("pdf(x)")

plt.show()
```
